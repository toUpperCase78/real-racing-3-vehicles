Well, for a more complex operation, I'd like to **combine these two criterions**: The count is performed by comparing **the classes against their purchase types**. To make it work, I have to create another column to store only the purchase type as string (e.g. only `Rdollar`, `Gold` and `Mdollar`).

Then, a pivot table will be useful here:

```python
def get_purchase_type(x):
    if type(x) != float and x[-7:] == 'Rdollar':    return 'Rdollar'
    elif type(x) != float and x[-4:] == 'Gold':     return 'Gold'
    elif type(x) != float and x[-7:] == 'Mdollar':  return 'Mdollar'
rr3_df['Purchase Type'] = rr3_df['Price'].apply(get_purchase_type)
```

```python
cls_pt_table = rr3_df.pivot_table(index=['Class'], columns=['Purchase Type'], values=['Model'], aggfunc='count').fillna(0)
cls_pt_table.astype(int)
```

**That's more like it!** That table gave us much more insights for _the count of classes against the purchase types of the vehicles!_ 

Compare this with the bar graphs above; do these numbers match when summed?

Meanwhile, this cross-counting technique reveals another fact: Vehicles that are purchasable with **M\$** are all from **R class!** Plus, **Gold purchase-type** is dominant among all classes.

Now, let's take a different perspective: Why not creating **ECDFs (Empirical Cumulative Distribution Functions)** with the values from vehicle stats?

```python
# The main function to create the ECDF
def ECDF(data):
    n = len(data)
    x = np.sort(data)
    y = np.arange(1, n+1) / n
    return x, y
plt.figure(figsize=(13,10))
plt.subplot(2,2,1)
x_speed, y_speed = ECDF(rr3_df['Top Speed'])
plt.scatter(x_speed, y_speed, marker='.', color='#cc8100', s=70, alpha=0.7)
plt.title('ECDF of Top Speed Values', fontsize=16, fontweight='bold')
plt.xlabel('Top Speed (kph)', fontsize=14, fontweight='bold')
plt.grid(alpha=0.4)
plt.subplot(2,2,2)
x_acc, y_acc = ECDF(rr3_df['Acceleration'])
plt.scatter(x_acc, y_acc, marker='.', color='#00ad68', s=70, alpha=0.7)
plt.title('ECDF of Acceleration Values', fontsize=16, fontweight='bold')
plt.xlabel('Acceleration (second)', fontsize=14, fontweight='bold')
plt.grid(alpha=0.4)
plt.subplot(2,2,3)
x_brake, y_brake = ECDF(rr3_df['Braking'])
plt.scatter(x_brake, y_brake, marker='.', color='#ad0048', s=70, alpha=0.7)
plt.title('ECDF of Braking Values', fontsize=16, fontweight='bold')
plt.xlabel('Braking (meters)', fontsize=14, fontweight='bold')
plt.grid(alpha=0.4)
plt.subplot(2,2,4)
x_grip, y_grip = ECDF(rr3_df['Grip'])
plt.scatter(x_grip, y_grip, marker='.', color='#7300ad', s=70, alpha=0.7)
plt.title('ECDF of Cornering Grip Values', fontsize=16, fontweight='bold')
plt.xlabel('Cornering Grip (g)', fontsize=14, fontweight='bold')
plt.grid(alpha=0.4)
plt.tight_layout()
plt.show()
```

Here are the results we can observe from the figure above:

* Most of the vehicles have the TOP SPEED of between **300 and 350 kph** (pay attention to the density of dots). Besides, in general, **over 80%** of the vehicles have **less than 350 kph** top speed.
* For ACCELERATION, the related dots rapidly rise up in the left side; **approximately 80%** of the vehicles in the game have **less than 4 seconds** of accelerations. Remarkably, some small groups of dots can appear vertically as they have the same acceleration values.
* Speaking of BRAKING values, again an almost-regular distribution of values can be seen, though some disconnections exist. Vehicles most likely have around **30 meters** of braking performance. Note the situation in the bottom left; that's because **all Formula 1 vehicles**, plus a few in the Road Collection, have very stong braking capabilities **(under 20 meters)**. In contrast, several vehicles have worse values than others **(which is over 40 meters)**.
* The distribution of values for CORNERING GRIP is somewhat similar to the ones in acceleration. Most vehicles have between **0.9-1.5 g** cornering capability and the skewness derived from the dots continues until **about 1.8 g**. Again, don't be surprised as **all Formula 1** and similar vehicles have **slightly more than 3.0 g**, thus the dots gathering around there consequently. Plus, an outlier exists whose value is exatly **4.0 g**! _If you'd like to know which one had this extreme value, don't hesitate to check out the previous analysis..._

Next, can we find any correlation with **PR** values if we combine these ECDFs above, through our eyes?

For this to work, all these values above, including PR, **must be normalized**; between 0.0 and 1.0! After that, we come across this output:

```python
x_pr, y_pr = ECDF(rr3_df['PR'])
x_pr /= x_pr.max()
x_speed = x_speed.astype('float64')
x_speed -= x_speed.min()
x_speed /= x_speed.max()
x_acc -= x_acc.min() 
x_acc /= x_acc.max()
x_brake -= x_brake.min()
x_brake /= x_brake.max()
x_grip -= x_grip.min()
x_grip /= x_grip.max()
plt.figure(figsize=(15,9))
plt.scatter(x_pr, y_pr, marker='.', color='#666666', s=400, label='PR', alpha=0.7)
plt.scatter(x_speed, y_speed, marker='.', color='#cc8100', s=60, label='Top Speed', alpha=0.4)
plt.scatter(x_acc, y_acc, marker='.', color='#00ad68', s=60, label='Acceleration', alpha=0.4)
plt.scatter(x_brake, y_brake, marker='.', color='#ad0048', s=60, label='Braking', alpha=0.4)
plt.scatter(x_grip, y_grip, marker='.', color='#7300ad', s=60, label='Grip', alpha=0.4)
plt.title('Combined ECDFs of All Stats Against PR', fontsize=20, fontweight='bold')
plt.xlabel('All Stat Values of All Vehicles (normalized)', fontsize=16, fontweight='bold')
plt.ylabel('ECDF', fontsize=16, fontweight='bold')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=13)
plt.grid(alpha=0.4, zorder=1)
plt.show()
```

Hmmm, after normalizing everything here, it might be weird at first!

Anyway, from this complicated ECDF plot, from **20%** up to **about 45%**, the PR value roughly follow the values of **top speed**, sometimes being exactly on them and sometimes staying close both at the left or right side. However, **after 45% to 90%**, the PR value becomes the rightmost one!

From the plot above additionally, **grip** and **acceleration** are on the left part, **from 10% to 50%**; **braking** and **top speed** are on the right part, **from about 5% to almost 100%**, plus they swap positions at about 19%.

While determining the PR, in my opinion, braking and top speed together have **stronger effects** while grip and acceleration have **weaker**. _All in all, it may not be a great method to completely imply the relationships, but it was a nice attempt to get it there, right?_

After the ECDF operation, now I'd like to see the **correlations between these 4 stat groups of values**... Let's tackle them all by creating plots individually and find their relationships in terms of _positivity_ or _negativity with a solid line:_

```python
top_speed = rr3_df['Top Speed'].values
acc = rr3_df['Acceleration'].values
braking = rr3_df['Braking'].values
grip = rr3_df['Grip'].values
plt.figure(figsize=(15.5,10))
plt.subplot(2,3,1)
z = np.polyfit(top_speed, acc, 1)
plt.plot(top_speed, acc, marker='d', linestyle='none', alpha=0.3, color='darkblue')
plt.plot([top_speed.min(), top_speed.max()], [z[1], z[1]+(z[0]*(top_speed.max()-top_speed.min()))], color='darkblue')
plt.xlabel('Top Speed (kph)', fontsize=14, fontweight='bold')
plt.ylabel('Acceleration (s)', fontsize=14, fontweight='bold')
plt.grid(alpha=0.3)
plt.subplot(2,3,2)
z = np.polyfit(top_speed, braking, 1)
plt.plot(top_speed, braking, marker='d', linestyle='none', alpha=0.3, color='red')
plt.plot([top_speed.min(), top_speed.max()], [z[1], z[1]+(z[0]*(top_speed.max()-top_speed.min()))], color='red')
plt.xlabel('Top Speed (kph)', fontsize=14, fontweight='bold')
plt.ylabel('Braking (m)', fontsize=14, fontweight='bold')
plt.grid(alpha=0.3)
plt.subplot(2,3,3)
z = np.polyfit(top_speed, grip, 1)
plt.plot(top_speed, grip, marker='d', linestyle='none', alpha=0.3, color='green')
plt.plot([top_speed.min(), top_speed.max()], [z[1], z[1]+(z[0]*(top_speed.max()-top_speed.min()))], color='green')
plt.xlabel('Top Speed (kph)', fontsize=14, fontweight='bold')
plt.ylabel('Grip (g)', fontsize=12, fontweight='bold')
plt.grid(alpha=0.3)
plt.subplot(2,3,4)
z = np.polyfit(acc, braking ,1)
plt.plot(acc, braking, marker='d', linestyle='none', alpha=0.3, color='darkorange')
plt.plot([acc.min(), acc.max()], [z[1], z[1]+(z[0]*(acc.max()-acc.min()))], color='darkorange')
plt.xlabel('Acceleration (s)', fontsize=14, fontweight='bold')
plt.ylabel('Braking (m)', fontsize=14, fontweight='bold')
plt.grid(alpha=0.3)
plt.subplot(2,3,5)
z = np.polyfit(acc, grip, 1)
plt.plot(acc, grip, marker='d', linestyle='none', alpha=0.3, color='purple')
plt.plot([acc.min(), acc.max()], [z[1], z[1]+(z[0]*(acc.max()-acc.min()))], color='purple')
plt.xlabel('Acceleration (s)', fontsize=14, fontweight='bold')
plt.ylabel('Grip (g)', fontsize=14, fontweight='bold')
plt.grid(alpha=0.3)
plt.subplot(2,3,6)
z = np.polyfit(braking, grip, 1)
plt.plot(braking, grip, marker='d', linestyle='none', alpha=0.3, color='black')
plt.plot([braking.min(), braking.max()], [z[1], z[1]+(z[0]*(braking.max()-braking.min()))], color='black')
plt.xlabel('Braking (m)', fontsize=14, fontweight='bold')
plt.ylabel('Grip (g)', fontsize=14, fontweight='bold')
plt.grid(alpha=0.3)
plt.show()
```

_**It is great to examine all these correlations in just one plotting group, isn't it?**_ 

At the same time, let's get the **correlation coefficients (Pearson)** in numerical way:

```python
tspeed_values = rr3_df['Top Speed'].values
acc_values = rr3_df['Acceleration'].values
brake_values = rr3_df['Braking'].values
grip_values = rr3_df['Grip'].values
print("** CORRELATIONS **")
print("Top Speed & Acceleration: {:6.3f}".format(np.corrcoef(tspeed_values, acc_values)[0,1]))
print("Top Speed & Braking:      {:6.3f}".format(np.corrcoef(tspeed_values, brake_values)[0,1]))
print("Top Speed & Grip:         {:6.3f}".format(np.corrcoef(tspeed_values, grip_values)[0,1]))
print("Acceleration & Braking:   {:6.3f}".format(np.corrcoef(acc_values, brake_values)[0,1]))
print("Acceleration & Grip:      {:6.3f}".format(np.corrcoef(acc_values, grip_values)[0,1]))
print("Braking & Grip:           {:6.3f}".format(np.corrcoef(brake_values, grip_values)[0,1]))
```

_**Compare both the values here and the corresponding plots, including the lines going upward or downward. Consequently, they mutually look very logical, yes?**_

It is definitely right to say that _Top Speed & Acceleration_ are **negatively correlated** with a magnitude of 0.551, with regards to the numerical values. That is, if the vehicle can reach **higher top speed**, it is expected to have **faster acceleration**, thus **lower values**. This is the claim we get...

Similarly, _Top Speed & Braking,_ _Acceleration & Grip_ and _Braking & Grip_ are also **negatively correlated** with different magnitudes as seen above. (Respective values are -0.288, -0.626, -0.803)

On the other hand, these two pairs are **positively correlated**: _Top Speed & Grip_ and _Acceleration & Braking_ with magnitude of 0.280 and 0.721, respectively. So, if one value is increasing, then it is expected for the other one to go up, where greater effect can be seen for the second pair.

Now, can we consolidate everything such that their correlations are packed together and color-coded? Seaborn's **heat map** will help us greatly for that! This time, we include three additional columns: **PR**, **Service Cost** and **Service Time**.

```python
rr3_df_vehicle_values = rr3_df[['Top Speed', 'Acceleration', 'Braking', 'Grip', 'PR', 'Service Cost', 'Service Time']]
plt.figure(figsize=(15,8))
sns.set(font_scale=1.25)
sns.heatmap(rr3_df_vehicle_values.corr(), cmap='YlOrRd', annot=True, vmin=-1, vmax=1, fmt='.3f')
plt.title('Real Racing 3 Correlations of Vehicle Values in Heatmap', fontsize=22, fontweight='bold')
plt.show()
```
